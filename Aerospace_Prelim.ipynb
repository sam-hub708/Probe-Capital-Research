```python
jpm = yf.Ticker("JGBL")

# get all stock info
jpm.info

# get historical market data
hist = jpm.history(period="max")

def numOfDays(date1, date2):
  #check which date is greater to avoid days output in -ve number
    if date2 > date1:
        return (date2-date1).days
    else:
        return (date1-date2).days
```

```python
assets = ['SPIR', 'ASTS', 'SPCE', 'PL', 'LUNR', 'LLAP', 'RKLB','TSAT', 'SATS', 'RDW', 'SIDU', 'BKSY', 'MDA.TO', 'LMT', 'NOC', 'GE', 'HWM', 'RTX']
start_date = date(2024, 1, 1)
end_date = date(2024, 11, 20)
duration = numOfDays(start_date, end_date)*252/365.25
df = yf.download(assets, start=start_date, end=end_date)
df = df['Adj Close']
```

```python
#returns_d1 = df[["^IRX","^TNX"]].diff().dropna()/100
returns_d = df.pct_change().dropna()
#frames = [returns_d1, returns_d2]
#returns_d = pd.concat(frames, axis=1)
returns_a = returns_d.sum()
returns_a.plot(kind='bar')
plt.title('Period Expected Return per Asset')
plt.show()
#returns_a
```

```python
vol = returns_d.std()*np.sqrt(duration)
#print(vol)
vol.plot(kind='bar')
plt.title('Period Volatility per Asset')
plt.show()
```

```python
cov_matrix = returns_d.cov()
sns.set(rc={'figure.figsize':(20, 10)})
sns.heatmap(returns_d.corr(),annot=True)
```

```python
n_assets = len(assets)
i = 1000
returns_p = []
vol_p = []
weights_p = []
for x in range(i):
  weights = np.random.random(n_assets)
  weights /= np.sum(weights)
  returns_p.append(np.sum(weights*returns_d.mean())*duration)
  vol_p.append(np.sqrt(np.dot(np.transpose(weights), np.dot(returns_d.cov()*duration, weights))))

returns_p = np.array(returns_p)
vol_p = np.array(vol_p)

portfolios = pd.DataFrame({'Return': returns_p, 'Vol': vol_p})
portfolios
```

```python
portfolios.plot(x='Vol', y='Return', kind='scatter', color='blue')
plt.title(f'Efficient Frontier | Iterations: 1000')
plt.xlabel('Expected Volatility')
plt.ylabel('Expected Return')
min_vol_p = portfolios.loc[portfolios['Vol'].idxmin()]
max_returns_p = portfolios.loc[portfolios['Return'].idxmax()]
plt.scatter(x=min_vol_p[1], y=min_vol_p[0], color='g', marker='*', s=200)
plt.scatter(x=max_returns_p[1], y=max_returns_p[0], color='r', marker='*', s=200)
plt.show()
print(f'The safest Portfolio will return: {round(100*min_vol_p[0],2)}%')
print(f'The Riskiest Portfolio will return: {round(100*max_returns_p[0],2)}%')
```
